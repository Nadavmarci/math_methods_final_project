```python
# Extract unique values from the first column
unique_countries = eigenvec_data['Country'].unique()

# Print the unique country names
print("Unique country names:", unique_countries)
```

```python
# Rename the first column to 'Country' instead of creating a duplicate column
eigenvec_data.columns = ['Country', 'Sample_ID', 'PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10', 'PC11', 'PC12', 'PC13', 'PC14', 'PC15', 'PC16', 'PC17', 'PC18', 'PC19', 'PC20']

country_lat_long = {
    'GBR': (51.5074, -0.1278),  # United Kingdom 
    'FIN': (60.1699, 24.9384),  # Finland 
    'IBS': (40.4637, -3.7492),  # Iberian Peninsula 
    'CEU': (48.8566, 2.3522),  # Central Europe 
    'TSI': (42.6043, 13.3714)  # Toscani in Italy 
}

# Adding latitude and longitude to the dataset
eigenvec_data['Lat'] = eigenvec_data['Country'].map(lambda x: country_lat_long[x][0])
eigenvec_data['Long'] = eigenvec_data['Country'].map(lambda x: country_lat_long[x][1])

# Displaying the first five rows after the update
print(eigenvec_data.head())
```

```python
import numpy as np
from scipy.stats import pearsonr
from scipy.optimize import minimize

# PCA points - PC1 and PC2
pc1 = eigenvec_data['PC1']
pc2 = eigenvec_data['PC2']

# Longitude and latitude values
longitudes = eigenvec_data['Long']
latitudes = eigenvec_data['Lat']

# Function to rotate axes
def rotate_vectors(theta, v1, v2):
    # Create a rotation matrix
    rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                                [np.sin(theta), np.cos(theta)]])
    
    # Rotate each vector separately
    rotated_v1 = np.cos(theta) * v1 - np.sin(theta) * v2  
    rotated_v2 = np.sin(theta) * v1 + np.cos(theta) * v2 
    
    return rotated_v1, rotated_v2  

# Objective function that maximizes the correlation between rotated axes and longitude/latitude
def objective_function(theta):
    # Rotate the axes
    rotated_pc1, rotated_pc2 = rotate_vectors(theta, pc1, pc2)
    
    # Compute the correlation between rotated PC1 and Longitude
    corr_pc1_long, _ = pearsonr(rotated_pc1, longitudes)
    
    # Compute the correlation between rotated PC2 and Latitude
    corr_pc2_lat, _ = pearsonr(rotated_pc2, latitudes)
    
    # Sum of correlations (as defined in the equation)
    return - (corr_pc1_long + corr_pc2_lat)  

# Find the angle that maximizes correlation
result = minimize(objective_function, 0, bounds=[(-np.pi, np.pi)])

# Optimal angle in radians
optimal_angle_rad = result.x[0]
# Convert the angle to degrees
optimal_angle_deg = np.degrees(optimal_angle_rad)
print(f"Optimal angle (degrees): {optimal_angle_deg}")
```

```python
# Rotate again using the optimal angle
rotated_pc1, rotated_pc2 = rotate_vectors(optimal_angle_rad, pc1, pc2)
```

```python
# Print the values of rotated_pc1 and rotated_pc2
print("rotated_pc1 values:", rotated_pc1)
print("rotated_pc2 values:", rotated_pc2)
```

```python
import seaborn as sns

# Create a scatter plot of rotated PC1 and PC2
plt.figure(figsize=(10, 8))

# Create a scatter plot with colors based on countries
sns.scatterplot(x=rotated_pc1, y=rotated_pc2, hue=eigenvec_data['Country'], palette='tab20', s=100, edgecolor='black')

# Add title and axis labels
plt.title("PCA: Rotated PC1 vs PC2 (Optimal Angle)", fontsize=15)
plt.xlabel("PC1 (Rotated)", fontsize=12)
plt.ylabel("PC2 (Rotated)", fontsize=12)

# Display the legend (color mapping for countries)
plt.legend(title='Countries', bbox_to_anchor=(1.05, 1), loc='upper left')

# Set axis limits to focus on the relevant range
plt.xlim(-0.2, 0.2)  
plt.ylim(-0.1, 0.2)  

# Show the grid and display the plot
plt.grid(True)
plt.show()
```
